```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-chinese')
```

```python
sentences = ['我爱逻辑','今天天气真好啊']
```

```python
inputs = tokenizer(sentences, padding=True,return_tensors='pt')
```
