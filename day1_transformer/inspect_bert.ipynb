```python
import torch
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
text = 'Logic is the foudation of AI.'
```

```python
tokens = tokenizer.tokenize(text)
```

```python
tokens
```

```python
ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
ids
```

```python
inputs = tokenizer(text, return_tensors="pt")
```

```python
inputs
```

```python
from transformers import AutoModel
```

```python
model = AutoModel.from_pretrained('bert-base-uncased')
```

```python
with torch .no_grad():
    outputs = model(**inputs)
```

```python
outputs
```

```python
last_hidden_states = outputs.last_hidden_state
```

```python
last_hidden_states.shape
```

```python
import torch.nn.functional as F
```

```python
words = ['King','Queen','Apple']
```

```python
inputs = tokenizer(words,return_tensors='pt',padding=True)
```

```python
inputs
```

```python
with torch.no_grad():
    outputs = model(**inputs)
```

```python
cls_embeddings = outputs.last_hidden_state[:,0,:]
```

```python
score_kq = F.cosine_similarity(cls_embeddings[0], cls_embeddings[1], dim=0)
```

```python
score_kq
```

```python
score_ka = F.cosine_similarity(cls_embeddings[0], cls_embeddings[2], dim=0)
```

```python
score_ka
```
