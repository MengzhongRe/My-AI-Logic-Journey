```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

```python
type(tokenizer.backend_tokenizer)
```

```python
tokenizer.backend_tokenizer.normalizer.normalize_str('Héllò hôw are ü?')
```
